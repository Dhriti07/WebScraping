```python
import requests
from bs4 import BeautifulSoup 
import json
```

```python
def makeSoupFromLink(link):
    fetched = False
    resp = None
    while fetched == False:
        try:
            resp = requests.get(link)
            fetched = True
        except:
            print("[DOWNLOAD ERROR] Trying again....")
    return BeautifulSoup(resp.content, 'html.parser')
```

```python
import pandas as pd
cols = ['link','name','website','city','state','zip','phone','cont_name','org']
df = pd.DataFrame(columns=cols)
for pageNum in range(250, 500):
  print("Page:" + str(pageNum))
  link = 'https://marketplace.akc.org/puppies?page=' + str(pageNum)
  soup = makeSoupFromLink(link)
  text = soup.main.script.get_text()
  text = text.replace("window.marketplace = ", "")
  text = text.replace('state:', '\"state\":')
  
  urlArr = json.loads(text)['state']['app']['search_results']['pages']['data']

  for elm in urlArr:
    link = "http://marketplace.akc.org" + elm['website']
    print(link)
    soup = makeSoupFromLink(link)
    text = soup.main.script.get_text()
    text = text.replace("window.marketplace = ", "")
    text = text.replace('state:', '\"state\":')

    obj = json.loads(text)['state']['app']
    breeds = obj['service_breeds']
    location = obj['location']
    name =[]
    website = location['external_website']
    zip_loc = location['zip_location']
    city = zip_loc['city']
    state = zip_loc['state']
    zip = zip_loc['zip']
    phone = location['phone']
    cont_name = location['contact_name']
    org = obj['organization']['name']
    for i in breeds:
      name.append(i['name'])
    Dict = {'link':link, 'name':name, 'website': website, 'city': city, 'state': state, 'zip':zip, 'phone':phone, 'cont_name':cont_name, 'org':org}
    df = df.append(Dict, ignore_index=True)
    df.to_csv('250to500.csv')
```
