```python
import pandas as pd 
```

```python
df = pd.read_csv('1to1168.csv')
```

```python
df
```

```python
df = df.iloc[: , 1:]
```

```python
df
```

```python
for i in range(df['name'].size):
  df['name'][i] = df['name'][i].strip('][').split(', ')
df = df.explode(df.columns[2])
```

```python
df = df.reset_index(drop=True)
```

```python
df
```

```python
import numpy as np
for i in range(df['website'].size):
  if pd.isnull(df['website'][i]):
    str1 = df['link'][i]
    pos = str1.find('?')
    if(pos != -1):
      str1 = str1[:pos]
      df['website'][i] = str1
    else :
      str1 = str1.split("breeder")
      str2 = str1[1].split("/")
      str2 = str2[1].split("/")
      str3 = str1[0] + "breeder/" + str2[0] + "/"
      df['website'][i] = str3
```

```python
df
```

```python
df = df.drop(['link','page'], axis = 1)
```

```python
df = df.drop_duplicates(keep= "first")
```

```python
df
```

```python
df = df.reset_index(drop=True)
df
```

```python
for i in range(df['name'].size):
    df['name'][i] = df['name'][i].strip('"')
    df['name'][i] = df['name'][i].strip('\'')
```

```python
df=df.sort_values(by = ['name','state','org'], ascending = [True, True, True], na_position = 'first')
df
```

```python
df = df.reset_index(drop=True)
df
```

```python
df.to_csv('Formatted.csv')
```

```python
file1 = open("Final_Text.txt","w")
file1.truncate(0)
breeds =[]
for i in range(df['org'].size):
  if( df['name'][i] not in breeds):
    breeds.append(df['name'][i])
    file1.write(df['name'][i] + '\n\n')
  file1.write(df['org'][i])
  file1.write('\n')
  file1.write('-  ' + 'Contact:' + df['cont_name'][i] +'\n')
  file1.write('-  ' + 'Location:' + df['city'][i]+ ', ' +df['state'][i] + ' '+ str(df['zip'][i]) + '\n')
  file1.write('-  ' + 'Phone:' + str(df['phone'][i])+ '\n')
  file1.write('-  ' + 'Website:' + df['website'][i]+ '\n')
  file1.write('-  ' + 'Member of: AKC\n\n')
file1.close() 
```
