```python
import pandas as pd 
import csv
```

```python
# Reading out web scraped Data of Good Dog Website from CSV to a DataFrame 
df = pd.read_csv('site_2.csv')
df.head()
```

```python
# Remove Extra unnamed index column from csv 
df = df.iloc[:, 1:]
df.head()
```

```python
# Divide location column into City and State columns
df_location = df['loc'].str.split(",",expand=True)
df_location = df_location.iloc[:, :-2]
df_location.rename(columns = {0:'city', 1:'state'}, inplace = True)
df_location
```

```python
# If the Only one entry exists that entry corresponds to the state entry 
for i in range(df_location['city'].size):
  if (df_location['state'][i] == None):
    df_location['state'][i] = df_location['city'][i]
    df_location['city'][i] = None
```

```python
df_location
```

```python
# Append the location data frame to the originial dataframe 
df = df.join(df_location)
```

```python
df
```

```python
# CSV File Codes Containing State Codes of all the states 
# Use the CSV file to make a state_map dictionary corresponding to all states and their codes 
with open('Codes.csv', mode='r') as infile:
  reader = csv.reader(infile)
  with open('Codes_new.csv', mode='w') as outfile:
    writer = csv.writer(outfile)
    state_map = {rows[0]:rows[1] for rows in reader}
```

```python
state_map
```

```python
# Replace all full values of state to their state codes to bring about uniformity 
for i in range(df['state'].size):
  if(not (df['state'][i] == None)):
    if df['state'][i] in state_map.keys():
      df['state'][i] = state_map[df['state'][i]]
```

```python
df
```

```python
# Sort the values first by breed followed by state code and then followed by the name of the breeder 
df = df.sort_values(by = ['breed','state','name'], ascending = [True, True, True], na_position = 'first')
df
```

```python
#Fix the indexing of the dataframe 
df = df.reset_index(drop=True)
df
```

```python
# Store the storted and formatted dataframe as csv 
df.to_csv('formatted_site2.csv')
```

```python
# Append it to a text file to match the aim of the task 
file1 = open("Site_2.txt","w")
file1.truncate(0)
breeds =[]
for i in range(df['name'].size):
  if( df['breed'][i] not in breeds):
    breeds.append(df['breed'][i])
    file1.write(df['breed'][i] + '\n\n')
  file1.write(df['name'][i])
  file1.write('\n')
  file1.write('-  ' + 'Contact:' + df['cont_name'][i] +'\n')
  file1.write('-  ' + 'Location:' + str(df['city'][i])+ ', ' + str(df['state'][i]) + '\n')
  file1.write('-  ' + 'Website:' + df['website'][i]+ '\n')
  file1.write('-  ' + 'Member of: Good Dog\n\n')
file1.close() 
```
