```python
import requests
from bs4 import BeautifulSoup 
import pandas as pd
import json
```

```python
def makeSoupFromLink(link):
    fetched = False
    resp = None
    while fetched == False:
        try:
            resp = requests.get(link)
            fetched = True
        except:
            print("[DOWNLOAD ERROR] Trying again....")
    return BeautifulSoup(resp.content, 'html.parser')
```

```python
for i in range(len(links_arr)):
  links_arr[i] = links_arr[i].replace("https://www.gooddog.com/" , "https://www.gooddog.com/api/breeds/")
```

```python
breeds = []
for i in range(len(links_arr)):
  breeds.append(links_arr[i].replace("https://www.gooddog.com/api/breeds/" , ""))
```

```python
breeds
```

```python
for i in range(len(breeds)):
  breeds[i] = breeds[i].capitalize()
  breeds[i] = breeds[i].replace("-", " ")
```

```python
breeds
```

```python
# Create a data frame of all the data scraped from the website 
cols = ['page', 'breed' ,'name','website','loc','cont_name']
df = pd.DataFrame(columns = cols)

for j in range(len(links_arr)):
  link_prefix = links_arr[j]
  print("New Link" + str(j) + ":" + link_prefix)

  breed = breeds[j];
  pageNum = 0

  while (True):

    pageNum = pageNum +1
    link = link_prefix + "/breeder_profiles/index_v2.json/?city=&distance_in_miles=500&non_standard=false&page=" + str(pageNum) +"&pbo=false&per_page=1000&postal_code=147005&state="
    res = requests.get(link)

    if(res.json()['data'] == []):
      break

    print("PageNum:" + str(pageNum))

    for i in range(len(res.json()['data'])):

      sub_url = res.json()['data'][i]['url']
      print("https://www.gooddog.com" + sub_url)
      id = res.json()['data'][i]['id']
      unit_link = 'https://www.gooddog.com/api/breeder_profiles/'+ str(id) +'/breeder_interview.json/?preview=false'
      unit_res = requests.get(unit_link)
      name = unit_res.json()['data']['name']
      breeder_name = unit_res.json()['data']['breeder_full_name']
      website = unit_res.json()['data']['website_url']

      if(website == None):
        website = "https://www.gooddog.com/" + sub_url

      location = res.json()['data'][i]['facility_location']
      Dict = {'page':pageNum,'breed':breed , 'name': name, 'website': website, 'loc': location, 'cont_name':breeder_name}
      df = df.append(Dict, ignore_index=True)
      df.to_csv('site_2.csv')
```
